```python
import matplotlib.pyplot as plt
%matplotlib inline  #u can see visual diagram in jupyter notebook itself
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y,c='g')  #size should be same of x and y
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
y=x*x
```

```python
#line plot
plt.plot(x,y,'ro',linestyle='dashed',linewidth=2,markersize=10) #r* means * come, bo means blue dot comes,r+ means + comes
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('2D Diagram')
```

```python
##CREATING SUBPLOTS

plt.subplot(2,2,1) #1st quadrant , 2,2,1 means 2 rows 2 column and 1st quadrant
plt.plot(x,y,'r--')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.subplot(2,2,2) #2nd quadrant
plt.plot(y,x,'g--')
plt.subplot(2,2,3)
plt.plot(x,y,'b--')
plt.subplot(2,2,4)
plt.plot(x,y,'y--')
```

```python
##BAR PLOTS

x1=[2,3,4,5]
y1=[11,16,19,20]

x2=[3,9,11,14]
y2=[6,15,7,9]

plt.bar(x1,y1,color='g')
plt.bar(x2,y2)
```

```python
##HISTOGRAMS = frequency on elmts

a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
```

```python
#BOXPLOT USING MATPLOTLIB
data= [np.random.normal(0,std,100) for std in range(1,4)]
data

#to see any outlier
plt.boxplot(data)
```

```python
##PIE CHART

# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
